1. Using preprocessing.scale() function

````python
from sklearn.datasets import load_iris
````

````python
from sklearn import preprocessing
data = load_iris()
 
# separate the independent(Y) or target and dependent(X) variables
X_data = data.data
target = data.target
 
# standardization of dependent variables
standard = preprocessing.scale(X_data)
print(standard)
````

2. Using StandardScaler() function

````python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
 
data = load_iris()
scale= StandardScaler()
 
# separate the independent(Y) or target and dependent(X) variables
X_data = data.data
target = data.target
 
# standardization of dependent(X) variables
scaled_data = scale.fit_transform(X_data) 
print(scaled_data)
````
